### 1) Discount

```python
import numpy as np

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
gamma_1 = 0.999
gamma_2 = 0.99
gamma_3 = 0.975
gamma_4 = 0.95
gamma_5 = 0.9

timestep = 0.1

nr_steps = 120

steps = np.arange(nr_steps)

discount_1 = np.power(gamma_1,steps)
discount_2 = np.power(gamma_2,steps)
discount_3 = np.power(gamma_3,steps)
discount_4 = np.power(gamma_4,steps)
discount_5 = np.power(gamma_5,steps)

plt.plot(steps*timestep,
         discount_1,
         c='g',
         label="gamma = 0.999")
plt.plot(steps*timestep,
         discount_2,
         c='m',
         label='gamma = 0.99')
plt.plot(steps*timestep,
         discount_3,
         c='r',
         label='gamma = 0.975')
plt.plot(steps*timestep,
         discount_4,
         c='b',
         label='gamma = 0.95')
plt.plot(steps*timestep,
         discount_5,
         c='y',
         label='gamma = 0.9')

plt.legend(loc='lower right')
plt.xlabel('time [s]')
plt.ylabel('Discount')
plt.grid()
plt.title('Discount')
```
