## A) Auswertung des Trainings
Auswertung der csv, die w√§hrend des Trainings erstellt worden ist.

```python
import glob
filenames_progress = glob.glob( path_ray_results + "/*/progress.csv")
print(filenames_progress)
```

```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import json

progressList = []

for file_ in filenames_progress:
    df = pd.read_csv(file_,index_col=None, header=0)
    progressList.append(df)
    
episode_reward_mean = []
reward_min = []
reward_max = []

for progress in progressList:
    episode_reward_mean.append(progress[['episode_reward_mean']].values)
    reward_min.append(progress[['episode_reward_min']].values)
    reward_max.append(progress[['episode_reward_max']].values)
    
for reward_mean in episode_reward_mean:
    plt.plot(reward_mean)

plt.xlabel('iteration')
plt.ylabel('mean reward')
plt.grid()
plt.title('Training PPO')
#plt.ylim(0,400)
plt.xlim(800,1000)
```

```python
for rMax in reward_max:
    plt.plot(rMax)
    print(max(rMax))
    
plt.xlim(800,1000)
plt.ylim(1050,1250)
```
