```python
class visualizeArguments:
    def __init__ (self, result_dir,run,checkpoint_num,render_mode,save_render,horizon,num_rollouts,emission_to_csv,evaluate,teamspirit_0,teamspirit_1):
        self.result_dir = result_dir
        self.run = run
        self.checkpoint_num = checkpoint_num
        self.render_mode = render_mode
        self.save_render = save_render
        self.horizon = horizon
        self.num_rollouts = num_rollouts
        self.emission_to_csv = emission_to_csv
        self.evaluate = evaluate
        self.teamspirit_0 = teamspirit_0
        self.teamspirit_1 = teamspirit_1
```

```python
def replay(teamspirit_0,teamspirit_1):
    ray.init(num_cpus=3)
    path_ray_results = "/home/thorsten/ray_results/IntersectionExample/"
    folder_results = "PPO_MultiAgentIntersectionEnv_sharedPolicy_TeamSpirit-v0_0_2019-03-28_22-09-07zsxx39e6"
    checkpoint_number = 140
    num_rollouts = 10

    result_dir = path_ray_results + folder_results
    run = False
    checkpoint_num = str(checkpoint_number)
    render_mode = "no_render"
    save_render = False
    horizon = False
    emission_to_csv = False
    evaluate = False

    args = visualizeArguments(result_dir,run,checkpoint_num,render_mode,save_render,horizon,num_rollouts,emission_to_csv,evaluate,teamspirit_0,teamspirit_1)
    visualizer_rllib(args)
    ray.shutdown()
```

```python
for ts_0 in range(-10,11):
    print(ts_0)
    for ts_1 in range(-10,11):
        replay(ts_0/10.0,ts_1/10.0)
```
