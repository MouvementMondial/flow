```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import json

import glob
```

```python
def createActionPlot(folder_logs, filename_save):
    filenames_log = glob.glob( folder_logs + "/log_*.csv")

    csvList = []

    for file_ in filenames_log:
        df = pd.read_csv(file_,index_col=None, header=0)
        csvList.append(df)

    actions_1 = []
    actions_2 = []
    veh_speeds_1 = []
    veh_speeds_2 = []
    veh_poss_1 = []
    veh_poss_2 = []
    curr_rewards_1 = []
    curr_rewards_2 = []
    times = []

    for csv in csvList:
        actions_1.append(csv[['action_acc_1']].values)
        actions_2.append(csv[['action_acc_2']].values)
        veh_speeds_1.append(csv[['veh_speed_1']].values)
        veh_speeds_2.append(csv[['veh_speed_2']].values)
        veh_poss_1.append(csv[['veh_pos_1']].values)
        veh_poss_2.append(csv[['veh_pos_2']].values)
        curr_rewards_1.append(csv[['curr_reward_1']].values)
        curr_rewards_2.append(csv[['curr_reward_2']].values)
        times.append(np.arange(csv[['action_acc_1']].values.size)/10)
   
    N_RA = 10
    pltAlpha = 0.15

    # print action
    for time, action_1 in zip(times,actions_1):
        plt.plot(time[:-1],
                 action_1[:-1],
                 c='b',
                 alpha=pltAlpha)
    for time, action_2 in zip(times,actions_2):
        plt.plot(time[:-1],
                 action_2[:-1],
                 c='r',
                 alpha=pltAlpha)

    # print action with rolling avg
    for time, action_1 in zip(times,actions_1):
        action_1_RA = np.convolve(action_1[:,0],np.ones((N_RA,))/N_RA,mode='valid')
        plt.plot(time[N_RA//2:-N_RA//2],
                 action_1_RA[:-1],
                 c='b')
    for time, action_2 in zip(times,actions_2):
        action_2_RA = np.convolve(action_2[:,0],np.ones((N_RA,))/N_RA,mode='valud')
        plt.plot(time[N_RA//2:-N_RA//2],
                 action_2_RA[:-1],
                 c='r')    
    
    plt.xlabel('time [s]')
    plt.ylabel('action_acc [m/s^2]')
    plt.grid()
    plt.ylim(-3.5,3.5)
    plt.xlim(-0.5,15)
    plt.title('Vergleich der Action beider Fahrzeuge (rolling Avg)')
    plt.savefig(filename_save)
    plt.clf()
```

```python
import re
def natural_key(string_):
    """See http://www.codinghorror.com/blog/archives/001018.html"""
    return [int(s) if s.isdigit() else s for s in re.split(r'(\d+)', string_)]
```

```python
path_ray_results = "/home/thorsten/ray_results/IntersectionExample/"
folder_results = "PPO_MultiAgentIntersectionEnv-v0_0_2019-02-21_11-07-01v5vhj7jt"

folder_logs = glob.glob( path_ray_results+folder_results + "/results_run_*")
folder_logs = sorted(folder_logs,key=natural_key)
print(folder_logs)
i = 1
for folder_ in folder_logs:
    filename_save = path_ray_results+folder_results+"/action_"+str(i)+".png"
    createActionPlot(folder_,filename_save)
    i=i+1
```
